```python
# Set the style
sns.set(style="whitegrid")

# Create the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['co2'], bins=30, kde=True, color='skyblue')

# Label the chart
plt.title('Distribution of Carbon Dioxide Emissions (CO2)', fontsize=16)
plt.xlabel('CO2 Emissions (g/mile)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Show the plot
plt.show();
```

```python
# Create histogram for UCity
plt.figure(figsize=(10, 6))
sns.histplot(df['UCity'], bins=30, kde=True, color='salmon')

# Label the chart
plt.title('Distribution of Urban Fuel Economy (UCity)', fontsize=16)
plt.xlabel('Urban Fuel Economy (mpg)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# YOUR CODE HERE

# Plot histogram for 'cylinders'
plt.figure(figsize=(10, 6))
sns.histplot(df['cylinders'], bins=range(int(df['cylinders'].min()), int(df['cylinders'].max()) + 2), color='mediumseagreen')

# Label the chart
plt.title('Distribution of Engine Cylinders', fontsize=16)
plt.xlabel('Number of Cylinders', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Set plot size and style
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0, linewidths=0.5, fmt=".2f")

# Title and axis adjustments
plt.title('Correlation Heatmap of Fuel Economy Dataset', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show();
```

**# Data intepretation**  
1.**Strong Positive Correlations (Values close to +1)**
These indicate that when one variable increases, the other also tends to increase.

City, UCity, Highway, UHighway, and Comb: These are all strongly correlated (> 0.9) with each other, meaning different measures of fuel economy move together.

feScore and ghgScore: Very strongly positively correlated with each other (0.99), which makes sense since both relate to vehicle efficiency.

co2 vs. cylinders/displ: Strong positive correlation (~0.85–0.86), indicating that larger engines produce more CO2 emissions.

**2.Strong Negative Correlations (Values close to -1)**
These indicate that as one variable increases, the other tends to decrease.

Fuel efficiency (city, highway, comb, etc.) vs. CO2: Very strong negative correlation (~-0.9), which shows that more efficient cars emit less CO2.

Fuel efficiency vs. cylinders and displacement (displ): Also strongly negative (~-0.7 to -0.8), indicating that larger engines are less fuel efficient.

feScore and ghgScore vs. engine size (displ/cylinders): Around -0.78 to -0.79, supporting that bigger engines are less environmentally friendly.

**3.Interesting Observations**
pv2 and pv4 (perhaps powertrain type indicators) have a negative correlation with each other (-0.67), which might imply mutual exclusivity.

Year has very weak correlations with most variables, suggesting technological improvements might not be linearly reflected year by year.
